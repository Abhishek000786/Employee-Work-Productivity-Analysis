```python
import seaborn as sns
import matplotlib.pyplot as plt

mean_scores = df.groupby('Employment_Type')[['Productivity_Score', 'Well_Being_Score']].mean().reset_index()
print(mean_scores)
# Set the figure size
plt.figure(figsize=(8, 4))

# Bar plot for Productivity Score
plt.subplot(1, 2, 1)
sns.barplot(data=mean_scores, x='Employment_Type', y='Productivity_Score')
plt.title('Mean Productivity Score by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Mean Productivity Score')

# Bar plot for Well-Being Score
plt.subplot(1, 2, 2)
sns.barplot(data=mean_scores, x='Employment_Type', y='Well_Being_Score')
plt.title('Mean Well-Being Score by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Mean Well-Being Score')

# Show plots
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
correlation_matrix = df[['Hours_Worked_Per_Week', 'Productivity_Score', 'Well_Being_Score']].corr()
print(correlation_matrix)
```

```python
# Set the figure size
plt.figure(figsize=(10, 5))

# Scatter plot for Hours Worked vs. Productivity Score
sns.scatterplot(data=df, x='Hours_Worked_Per_Week', y='Productivity_Score', hue='Employment_Type', alpha=0.7)
plt.title('Hours Worked Per Week vs. Productivity Score')
plt.xlabel('Hours Worked Per Week')
plt.ylabel('Productivity Score')
#plt.grid(True)
plt.show()
```

```python
# Set the figure size
plt.figure(figsize=(10, 5))

# Scatter plot for Hours Worked vs. Well-Being Score
sns.scatterplot(data=df, x='Hours_Worked_Per_Week', y='Well_Being_Score', hue='Employment_Type', alpha=0.7)
plt.title('Hours Worked Per Week vs. Well-Being Score')
plt.xlabel('Hours Worked Per Week')
plt.ylabel('Well-Being Score')
plt.grid(True)
plt.show()
```
